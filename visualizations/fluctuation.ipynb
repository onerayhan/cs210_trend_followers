```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('merged_data.csv')

# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Extract the year from the dates
df['Year'] = df['Date'].dt.year

# Define the years of interest
years_of_interest = [2020, 2021, 2022, 2023]

# Create an empty list to store the fluctuation values for each year
fluctuation_values = []

# Perform time-series analysis and collect the fluctuation values for each year
for year in years_of_interest:
    # Filter the data for the current year
    df_year = df[df['Year'] == year]
    
    # Calculate the fluctuation between predicted and true values
    fluctuation = df_year['Predicted Value'] - df_year['True Value']
    
    # Append the fluctuation values to the list
    fluctuation_values.append(fluctuation)

# Plot the histogram of fluctuation values
plt.figure(figsize=(12, 6))
plt.hist(fluctuation_values, bins=20, label=years_of_interest)
plt.xlabel('Fluctuation')
plt.ylabel('Frequency')
plt.title('Histogram of Fluctuation Values')
plt.legend()
plt.grid(True)

# Show the histogram
plt.show()
```
